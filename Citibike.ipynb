```python
import pandas as pd
```

```python
df7 = pd.read_csv('Resources/201807-citibike-tripdata.csv')
df8 = pd.read_csv('Resources/201808-citibike-tripdata.csv')
df9 = pd.read_csv('Resources/201809-citibike-tripdata.csv')
```

```python
summer_df=pd.concat([df7,df8,df9])
```

```python
summer_df[["gender"]]=summer_df[["gender"]].replace(to_replace =1, 
                 value ="Male")
summer_df[["gender"]]=summer_df[["gender"]].replace(to_replace =2, 
                 value ="Female")
summer_df[["gender"]]=summer_df[["gender"]].replace(to_replace =0, 
                 value ="Unknown")
```
